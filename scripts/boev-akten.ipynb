Im ersten Schritt wird geschaut, ob sich mit den Namen aus den Rohdaten GND-Matches erzeugen lassen. Es wird eine neue Spalte 'matches' erzeugt, in der eine Liste von Dictionaries für alle Matchkandidaten festhält, welches Kriterium genau übereinstimmte und welches nicht. Erst im nächsten Schritt werden diese Dictionaries dann ausgewertet, so dass die Kriterien noch feingetunet werden können, die schließlich zu einer IDN führen, die als definitiver Match angesehen wird.

```python
import unicodedata
base_url = 'https://services.dnb.de/sru/authorities?'
params = {'recordSchema' : 'MARC21-xml',
          'operation': 'searchRetrieve',
          'version': '1.1',
          'maximumRecords': '100',
          'accessToken': '8090beb04b8a8ce79c2da2c1ee6890b8',
         }

def gnd_abfrage(row):
    print(f'{row.Nachname}, {row.Vorname}')
    params.update({'query': f'PER = {row.Nachname}, {row.Vorname} AND BBG=Tp*'})
    response = requests.get(base_url, params=params)
    response_xml = soup(response.content)
    
    try:
        matches = []
        for record in response_xml.find_all('record', {'type':'Authority'}):
            record_match = {}
            #idn bestimmen
            try:
                record_match['idn'] = record.find('controlfield', {'tag': '001'}).string.strip()
            except:
                record_match['idn'] = 'XX'

            #datum checken    
            try:
                datl = record.find('datafield', {'tag':'400'}).find('subfield', {'code':'d'}).get_text(strip=True).strip().split('-')
                #print(datl)
            except:
                datl = '0000-'

            #setzen von datum_match, wenn geburtsjahre übereinstimmen
            record_match['datum_match'] = True if str(row.geburt)[-4:] == datl[0] else False

            #liste der berufe aus gnd-satz auslesen
            berufsliste = []
            try:
                berufsfelder = record.find_all('datafield', {'tag':'550'})
                for item in berufsfelder:
                    beruf = item.find('subfield', {'code':'a'}).get_text(strip=True).strip()
                    beruf = unicodedata.normalize("NFC", aberuf)
                    berufsliste.append(beruf)
            except:
                berufsliste.append('0')
            #print(berufsliste)

            #setzen von beruf_match, wenn einschlägige berufe in gnd-berufsliste
            if 'Buchhändler' in berufsliste or 'Buchhändlerin' in berufsliste or 'Verleger' in berufsliste or 'Verlegerin' in berufsliste:
                record_match['beruf_match'] = True
            else:
                record_match['beruf_match'] = False    

            #wirkungsort bestimmen
            try:
                ort = record.find('datafield', {'tag':'551'}).find('subfield', {'code':'a'}).get_text(strip=True).strip()
                ort = unicodedata.normalize("NFC",ort)
            except:
                ort = '0'

            record_match['ort_match'] = True if str(row.Ort).strip() == ort else False
            matches.append(record_match)
    except:
        matches = []
    
    #print(matches)
    return matches
        
df['matches'] = df.apply(gnd_abfrage, axis=1)
```

# GND-Auswertung

Nun werden die Match-Kandidaten ausgewertet. Hier wurde ein Kandidat dann als Treffer gewertet, wenn mindestens zwei der Match-Kriterien true waren.

```python
#es wird gezählt, wie oft im dictionary in matches True steht. wenn größer gleich 2, wird die idn zurückgegeben

def gnd_auswertung(matches):
    try:
        idns = []
        for match in matches:
            treffer = 0
            for v in match.values():
                if isinstance(v, bool):
                    treffer += v
            if treffer >= 2:
                idns.append(match['idn'])
            else:
                None
    except:
        None
    return idns

df['idns'] = df.matches.apply(gnd_auswertung)
#df.matches.apply(gnd_auswertung)
    
```
