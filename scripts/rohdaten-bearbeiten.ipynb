```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HABV_91_1_v1.csv', sep=';', index_col='Id')
# df
```

```python
def lebensdaten(daten):
    try:
        daten_split = daten.split('-')
    except:
        return [float("NAN"),float("NAN")]
    if len(daten_split) == 2:
        if len(daten_split[0]) == 0 and len(daten_split[1]) == 10:
            return [float("NAN"),daten_split[1]]
        elif len(daten_split[0]) == 10 and len(daten_split[1]) == 0:
            return [daten_split[0], float("NAN")]
        elif len(daten_split[0]) == 10 and len(daten_split[1]) == 10:
            return [daten_split[0], daten_split[1]]
        else: return [float("NAN"),float("NAN")]
    else: return [float("NAN"),float("NAN")]

```

```python
# df["geburt", "tod"] = df.Lebensdaten.apply(lebensdaten, result_type ='expand', )
# df['p1'], df['p2'], df['p3'], df['p4'], df['p5'], df['p6'] = zip(*df['num'].map(powers))
df["geburt"], df["tod"] = zip(*df.Lebensdaten.map(lebensdaten))
```

```python
# df[['Lebensdaten','geburt','tod']].sample(50)
df.to_csv('HABV_91_1_v2.csv')
```

```python
def firmenart(row):
    if str(row.Sortiment).strip() == 'x':
        firmenart = 'Sortiment'
    elif str(row.Sortiment).strip() == 'zwibu':
        firmenart = 'Zwischenbuchhandel'
    elif str(row.Sortiment).strip() == 'vv' or row.Verlag == 'vv':
        firmenart = 'Verlagsvertreter'
    elif str(row.Verlag).strip() == 'x':
        firmenart = 'Verlag'
    else:
        firmenart = None
    return firmenart

df['firmenart'] = df[['Sortiment', 'Verlag']].apply(firmenart, axis=1)
df
        
```

```python
df.to_csv('HABV_91_1_extended_neu_gender_lebensdaten_matches_idns_firmenart.csv', sep=';')
```
